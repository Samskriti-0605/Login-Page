```python
woman=titanic[titanic["Sex"] =="female"]
```

```python
woman
```

```python
woman_Survived=woman[woman["Survived"]==1]
woman_not_Survived=woman[woman["Survived"]==0]
```

```python
man=titanic[titanic["Sex"] =="male"]
man_Survived=man[man["Survived"]==1]
man_not_Survived=man[man["Survived"]==0]
```

```python
man_not_Survived
```

```python
# Age - 1-100 --> 1,20 - bin1, 20-40-bin2, 40-60-bin3, 60-80-bin4, 80-100-bin5
# Histogram
sea_lib.histplot(woman_Survived.Age, bins=5, label="woman_Survived", kde=False)
sea_lib.histplot(woman_not_Survived.Age, bins=5, label="woman_not_Survived", kde=False)
plot_lib.legend()
```

```python
# Age - 1-100 --> 1,20 - bin1, 20-40-bin2, 40-60-bin3, 60-80-bin4, 80-100-bin5
# Histogram
sea_lib.histplot(man_Survived.Age, bins=5, label="man_Survived", kde=False)
sea_lib.histplot(man_not_Survived.Age, bins=5, label="man_not_Survived", kde=False)
plot_lib.legend()
```
